```python
#looking for the longest strings

strings = ["apple", "banana", "orange", "grape"]

long_strings = []

for string in strings:

  if len(string) > 5:
    long_strings.append(string)
    print(long_strings)

print("Long strings:", long_strings)
```

```python
#Finding the sum of numbers until the user enters 0

total = 0
number = int(input("Enter a number (0 to stop): "))

while number != 0:
  total += number
  number = int(input("Enter a number (0 to stop): "))
print("The sum of the numbers is:", total)


```

```python
# Guessing a number until the correct guess
secret_number = 7
guess = int(input("Guess a number between 1 and 10: "))
while guess != secret_number:
  if guess < secret_number:
    print("Too low!")
  else:
    print("Too high!")
  guess = int(input("Guess again: "))
print("Congratulations! You guessed the number.")

```

```python
# Iterating through a list until a specific element is found

my_list = [1, 2, 3, 4, 5, 6]
index = 0

while index < len(my_list):
  if my_list[index] == 4:
    print("Element 4 found at index", index)
    break  #change with continue
  index += 1
```

```python
# dictionary with while loop.

student_grades = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "David": 95,
    "Eve": 88
}

names = list(student_grades.keys())
i = 0
while i < len(names):
  name = names[i]
  grade = student_grades[name]
  # print(f"{name}: {grade}")  #pay attention to the print
  print(name,":",grade)
  i += 1
```
